```python
import xml.etree.ElementTree as ET
import json
import pandas as pd
from collections import defaultdict
```

```python
with open('british-diplomats-directory.xml') as fd:
    tree = ET.parse(fd)
root = tree.getroot()
```

```python
# there is a problem preventing using findall from the root branch.
# getting the list of all sections
sections = defaultdict(list)
section_attrib = '{http://www.w3.org/XML/1998/namespace}id'
for child in root.iter():
    if section_attrib in child.attrib:
        sections[child.attrib[section_attrib]].append(child)
```

```python
#section_b and section_c
section_b = dict()
for post_section in [post for entry in sections["section-b"] for post in entry.findall('.//persName/../../..')]:
    post_title = post_section.find("head").text
    section_b[post_title] =list()
    for person in post_section.findall(".//persName/.."):
        new_post = dict()
        new_post['persName'] = person.find("persName").text
        new_post['date'] = [a.text for a in person.findall("date")]
        new_post['roleName'] = [a.text for a in person.findall("roleName")]
        section_b[post_title].append(new_post)
```

```python
with open('section_b.json', 'w') as fp:
    json.dump(section_b, fp)
```

```python
#section_b and section_c
section_c = dict()
for post_section in [post for entry in sections["section-c"] for post in entry.findall('.//persName/../../..')]:
    post_title = post_section.find("head").text
    section_c[post_title] =list()
    for person in post_section.findall(".//persName/.."):
        new_post = dict()
        new_post['persName'] = person.find("persName").text
        new_post['date'] = [a.text for a in person.findall("date")]
        new_post['roleName'] = [a.text for a in person.findall("roleName")]
        section_c[post_title].append(new_post)
```

```python
with open('section_c.json', 'w') as fp:
    json.dump(section_c, fp)
```

```python
persons_clean = list()
for person in persons:
    ambassador = dict()
    pers_splt = person["persName"].split(",")
    ambassador["Name"] = (" ".join(pers_splt[1:])+" "+pers_splt[0]).title()
    ambassador["Date"] = person.get("date")
    ambassador["Blurb"] = ", ".join(" ".join(elt['rs'])+" "+"|".join(elt["dates"]) for elt in person['seg'])
    ambassador["persName"] = person["persName"]
    persons_clean.append(ambassador)
```
